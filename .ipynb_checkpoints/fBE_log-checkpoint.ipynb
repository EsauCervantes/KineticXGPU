```python
import numpy as np
from scipy.interpolate import PchipInterpolator

def make_loglog_p2f_interpolator(p, p2f, floor=1e-300):
    """
    Interpolates y(p)=p^2 f(p) in log-log space using PCHIP.
    Returns a callable y_of_p(p_new).
    """
    p = np.asarray(p, dtype=float)
    p2f = np.asarray(p2f, dtype=float)

    mask = (p > 0) & (p2f > 0)
    p_use = p[mask]
    y_use = p2f[mask]

    if len(p_use) < 2:
        raise ValueError("Need at least two valid positive points.")

    logp = np.log(p_use)
    logy = np.log(np.maximum(y_use, floor))

    interp_logy = PchipInterpolator(logp, logy, extrapolate=False)

    p_min = p_use[0]
    p_max = p_use[-1]

    def y_of_p(p_new):
        p_arr = np.asarray(p_new, dtype=float)
        scalar_input = np.ndim(p_arr) == 0
        p_arr = np.atleast_1d(p_arr)

        out = np.zeros_like(p_arr)
        inside = (p_arr >= p_min) & (p_arr <= p_max) & (p_arr > 0)

        out[inside] = np.exp(interp_logy(np.log(p_arr[inside])))

        if scalar_input:
            return float(out[0])
        return out

    return y_of_p
```

```python
import numpy as np

def compute_n_rho_from_p2f_interpolator(y_of_p, pmin, pmax, m, npts=2000):
    """
    Compute quantities proportional to

        n   ~ ∫ dp   p^2 f(p)
        rho ~ ∫ dp E p^2 f(p)

    where y_of_p(p) = p^2 f(p).

    Returns:
        n_val, rho_val
    """
    p_grid = np.logspace(np.log10(pmin), np.log10(pmax), npts)
    y_grid = y_of_p(p_grid)  # this is p^2 f
    E_grid = np.sqrt(p_grid**2 + m**2)

    n_val = np.trapz(y_grid, p_grid)
    rho_val = np.trapz(E_grid * y_grid, p_grid)

    return n_val, rho_val
```

```python
from scipy.special import kn

def infer_T_z_from_interpolator(y_of_p, pmin, pmax, m, npts=3000):
    p_grid = np.logspace(np.log10(pmin), np.log10(pmax), npts)
    y_grid = y_of_p(p_grid)   # y = p^2 f
    E_grid = np.sqrt(p_grid**2 + m**2)

    n = np.trapz(y_grid, p_grid) / (2.0 * np.pi**2)
    T = np.trapz(y_grid * (p_grid**2 / (3.0 * E_grid)), p_grid) / max(np.trapz(y_grid, p_grid), 1e-300)
    neq = (m**2 * T * kn(2, m / T)) / (2.0 * np.pi**2)
    z = n / max(neq, 1e-300)
    return T, z, n, neq

pmin = float(np.min(p))
pmax = float(np.max(p))

T_self, z_self, n_self, neq_self = infer_T_z_from_interpolator(my_int_self, pmin, pmax, mchi)
T_no_self, z_no_self, n_no_self, neq_no_self = infer_T_z_from_interpolator(my_int_no_self, pmin, pmax, m_chi)

print(f"SELF:    T = {T_self:.6e}, z = {z_self:.6e}, n = {n_self:.6e}, neq = {neq_self:.6e}")
print(f"NO SELF: T = {T_no_self:.6e}, z = {z_no_self:.6e}, n = {n_no_self:.6e}, neq = {neq_no_self:.6e}")

#T_self=1.48
#z_self=2.64e-3

# ---------- build inferred MB curves on your p grid ----------
E = np.sqrt(p**2 + m_chi**2)

fMB_self = z_self * np.exp(-E / T_self)
fMB_no_self = z_no_self * np.exp(-E / T_no_self)

p2fMB_self = p**2 * fMB_self
p2fMB_no_self = p**2 * fMB_no_self

# ---------- plot numerical + inferred MB ----------
plt.figure(figsize=(8,5))
plt.plot(p, E*p2f_self, label="numerical self")
plt.plot(p, E*p2f_no_self, label="numerical no self")
plt.plot(p, E*p2fMB_self, "--", label=fr"MB self ($T={T_self:.3e}$, $z={z_self:.3e}$)")
plt.plot(p, E*p2fMB_no_self, "--", label=fr"MB no self ($T={T_no_self:.3e}$, $z={z_no_self:.3e}$)")

plt.xlabel("p")
plt.ylabel(r"$p^2 f(p)$")
plt.xscale("log")
plt.yscale("linear")
plt.title(r"$p^2 f(p)$ at final $a$ with inferred MB comparison")
plt.grid(True, which="both")
plt.legend()
plt.tight_layout()
plt.show()

# ---------- ratio to MB ----------
plt.figure(figsize=(8,5))
plt.plot(p, p2f_self / np.maximum(p2fMB_self, 1e-300), label="self / MB")
plt.plot(p, p2f_no_self / np.maximum(p2fMB_no_self, 1e-300), label="no self / MB")
plt.axhline(1.0, linestyle="--")

plt.xlabel("p")
plt.ylabel(r"$(p^2 f)_{\rm num}/(p^2 f)_{\rm MB}$")
plt.xscale("log")
plt.grid(True, which="both")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from scipy.special import kn
import numpy as np
import matplotlib.pyplot as plt
from thermodynamics import N_eq

def compute_moments_from_interpolator(y_of_p, pmin, pmax, m, npts=3000):
    p_grid = np.logspace(np.log10(pmin), np.log10(pmax), npts)
    y_grid = y_of_p(p_grid)   # y = p^2 f(p)
    E_grid = np.sqrt(p_grid**2 + m**2)
    v_grid = p_grid / E_grid

    pref = 1.0 / (2.0 * np.pi**2)

    # basic densities
    n = pref * np.trapz(y_grid, p_grid)
    rho = pref * np.trapz(y_grid * E_grid, p_grid)
    P = pref * np.trapz(y_grid * (p_grid**2 / (3.0 * E_grid)), p_grid)

    # equation of state
    w = P / max(rho, 1e-300)

    # effective kinetic temperature proxy
    Tkin = P / max(n, 1e-300)

    # number-weighted averages
    avg_E = rho / max(n, 1e-300)
    avg_p = pref * np.trapz(y_grid * p_grid, p_grid) / max(n, 1e-300)

    avg_v = pref * np.trapz(y_grid * v_grid, p_grid) / max(n, 1e-300)
    avg_v2 = pref * np.trapz(y_grid * v_grid**2, p_grid) / max(n, 1e-300)
    v_rms = np.sqrt(max(avg_v2, 0.0))

    avg_p2_over_E2 = (
        pref * np.trapz(y_grid * (p_grid**2 / E_grid**2), p_grid)
        / max(n, 1e-300)
    )
    avg_p2_over_3E = (
        pref * np.trapz(y_grid * (p_grid**2 / (3.0 * E_grid)), p_grid)
        / max(n, 1e-300)
    )

    # MB-equivalent fugacity using Tkin only as a proxy
    neq_MB = (m**2 * Tkin * kn(2, m / max(Tkin, 1e-300))) / (2.0 * np.pi**2)
    z_MB = n / max(neq_MB, 1e-300)

    return {
        "n": n,
        "rho": rho,
        "P": P,
        "w": w,
        "Tkin": Tkin,
        "avg_E": avg_E,
        "avg_p": avg_p,
        "avg_v": avg_v,
        "avg_v2": avg_v2,
        "v_rms": v_rms,
        "avg_p2_over_E2": avg_p2_over_E2,
        "avg_p2_over_3E": avg_p2_over_3E,
        "neq_MB": neq_MB,
        "z_MB": z_MB,
        "p_grid": p_grid,
        "y_grid": y_grid,
    }

import numpy as np
from scipy.special import kn


def compute_moments_from_cbe(a, m, Ns, Ts):
    """
    Compute cBE moments directly from the cBE outputs:
      Ns(a): comoving number density N = a^3 n
      Ts(a): temperature parameter T
    """
    T = float(Ts(a))
    N = float(Ns(a))
    n = N / a**3

    x = m / max(T, 1e-300)

    # physical equilibrium number density
    n_eq = (m**2 * T * kn(2, x)) / (2.0 * np.pi**2)

    # fugacity
    z = n / max(n_eq, 1e-300)

    # MB pressure
    P = n * T

    # MB energy density
    K1 = kn(1, x)
    K2 = kn(2, x)
    rho = n * (3.0 * T + m * K1 / max(K2, 1e-300))

    w = P / max(rho, 1e-300)
    Tkin = P / max(n, 1e-300)
    avg_E = rho / max(n, 1e-300)

    return {
        "T": T,
        "N": N,
        "n": n,
        "n_eq": n_eq,
        "z": z,
        "rho": rho,
        "P": P,
        "w": w,
        "Tkin": Tkin,
        "avg_E": avg_E,
    }


def compute_velocity_moments_from_cbe(a, m, Ns, Ts, pmin, pmax, npts=3000):
    T = float(Ts(a))
    N = float(Ns(a))
    n = N / a**3

    x = m / max(T, 1e-300)
    n_eq = (m**2 * T * kn(2, x)) / (2.0 * np.pi**2)
    z = n / max(n_eq, 1e-300)

    p_grid = np.logspace(np.log10(pmin), np.log10(pmax), npts)
    E_grid = np.sqrt(p_grid**2 + m**2)
    v_grid = p_grid / E_grid

    f_grid = z * np.exp(-E_grid / max(T, 1e-300))
    y_grid = p_grid**2 * f_grid

    pref = 1.0 / (2.0 * np.pi**2)

    n_check = pref * np.trapz(y_grid, p_grid)
    avg_v = pref * np.trapz(y_grid * v_grid, p_grid) / max(n_check, 1e-300)
    avg_v2 = pref * np.trapz(y_grid * v_grid**2, p_grid) / max(n_check, 1e-300)
    v_rms = np.sqrt(max(avg_v2, 0.0))

    return {
        "T": T,
        "N": N,
        "n": n,
        "n_eq": n_eq,
        "z": z,
        "avg_v": avg_v,
        "avg_v2": avg_v2,
        "v_rms": v_rms,
    }
```

```python
import numpy as np
import torch
import matplotlib.pyplot as plt
from thermodynamics import N_eq
from collision_log import C_self_torch_logq

# --------------------------------------------------
# 1. Build the cBE MB target at fixed a = a_star
# --------------------------------------------------
a_star = float(5*10**4)
mchi = float(m_chi)

lam = 10**(-6)

q_np = q.detach().cpu().numpy()
p_np = q_np / a_star
E_np = np.sqrt(p_np**2 + mchi**2)

T_star = float(Ts(a_star))
N_star = float(Ns(a_star))
Neq_star = float(N_eq(a_star, T_star, mchi))
z_star = N_star / max(Neq_star, 1e-300)

f_MB_target_np = z_star * np.exp(-E_np / T_star)

print(f"a*     = {a_star:.6e}")
print(f"T_cBE  = {T_star:.6e}")
print(f"N_cBE  = {N_star:.6e}")
print(f"Neq    = {Neq_star:.6e}")
print(f"z_cBE  = {z_star:.6e}")

# --------------------------------------------------
# 2. Build a small perturbation around MB
#    Example: smooth log-tilt
# --------------------------------------------------
p0 = np.sqrt(p_np.min() * p_np.max())
eps = 0.05

shape = np.log(p_np / p0)
shape = shape - np.average(shape, weights=np.maximum(f_MB_target_np, 1e-300))

deltaf_np = eps * f_MB_target_np * shape

# Keep positivity
f_test_np = f_MB_target_np + deltaf_np
f_test_np = np.maximum(f_test_np, 1e-300)
deltaf_np = f_test_np - f_MB_target_np

# Move to torch
f_test_t = torch.tensor(f_test_np, device=device, dtype=dtype)

# --------------------------------------------------
# 3. Evaluate raw self-collision operator
# --------------------------------------------------
C_self_t, E_t, p_t, dp_t, diag = C_self_torch_logq(
    f=f_test_t,
    a=a_star,
    q=q,
    logq0=logq0,
    dlogq=dlogq,
    log_space=log_space,
    m=m_chi,
    lam=lam,
    Ng=64,
    batch_size=batch_size,
    apply_feedback_projection=False,
    return_diagnostics=True,
)

C_self_np = C_self_t.detach().cpu().numpy()
p_np = p_t.detach().cpu().numpy()
E_np = E_t.detach().cpu().numpy()
dp_np = dp_t.detach().cpu().numpy()

print("diag =", diag)

# --------------------------------------------------
# 4. Estimate local relaxation rate Gamma_eff(p)
#
#    If C ~ -Gamma * deltaf, then
#      Gamma_eff = -C/deltaf
# --------------------------------------------------
Gamma_loc_np = -C_self_np / np.maximum(np.abs(deltaf_np), 1e-300)

# bins where the diagnostic is meaningful
mask = (
    (f_MB_target_np > 1e-12 * np.max(f_MB_target_np)) &
    (np.abs(deltaf_np) > 1e-12 * np.max(np.abs(deltaf_np))) &
    np.isfinite(Gamma_loc_np)
)

# positive damping bins only
mask_pos = mask & (Gamma_loc_np > 0.0)

# robust summaries
if np.any(mask_pos):
    Gamma_eff_max = np.max(Gamma_loc_np[mask_pos])
    Gamma_eff_med = np.median(Gamma_loc_np[mask_pos])
    Gamma_eff_p90 = np.percentile(Gamma_loc_np[mask_pos], 90)
else:
    Gamma_eff_max = np.nan
    Gamma_eff_med = np.nan
    Gamma_eff_p90 = np.nan

H_star = float(cosmo.H_of_a(a_star))

print(f"H(a*)                 = {H_star:.6e}")
print(f"Gamma_eff max         = {Gamma_eff_max:.6e}")
print(f"Gamma_eff median      = {Gamma_eff_med:.6e}")
print(f"Gamma_eff p90         = {Gamma_eff_p90:.6e}")
print(f"Gamma_eff max / H     = {Gamma_eff_max / H_star:.6e}")
print(f"Gamma_eff median / H  = {Gamma_eff_med / H_star:.6e}")
print(f"Gamma_eff p90 / H     = {Gamma_eff_p90 / H_star:.6e}")

# --------------------------------------------------
# 5. Optional weighted RMS estimate
# --------------------------------------------------
w_np = p_np**2 * dp_np
if np.any(mask):
    num = np.sum(w_np[mask] * np.abs(C_self_np[mask]))
    den = np.sum(w_np[mask] * np.abs(deltaf_np[mask]))
    Gamma_eff_rms = num / max(den, 1e-300)
else:
    Gamma_eff_rms = np.nan

print(f"Gamma_eff RMS         = {Gamma_eff_rms:.6e}")
print(f"Gamma_eff RMS / H     = {Gamma_eff_rms / H_star:.6e}")

# --------------------------------------------------
# 6. Plots
# --------------------------------------------------
plt.figure(figsize=(8,5))
plt.plot(p_np, deltaf_np, label=r"$\delta f$")
plt.xscale("log")
plt.xlabel("p")
plt.ylabel(r"$\delta f$")
plt.title("Perturbation around MB target")
plt.grid(True, which="both")
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,5))
plt.plot(p_np, C_self_np, label=r"$C_{\rm self}[f_{\rm test}]$")
plt.xscale("log")
plt.axhline(0.0, ls="--")
plt.xlabel("p")
plt.ylabel(r"$C_{\rm self}$")
plt.title("Raw self-collision response")
plt.grid(True, which="both")
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,5))
plt.plot(p_np[mask], Gamma_loc_np[mask], ".", ms=4, label=r"$\Gamma_{\rm loc}(p)$")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("p")
plt.ylabel(r"$\Gamma_{\rm eff}(p)$")
plt.title("Local linearized relaxation rate")
plt.grid(True, which="both")
plt.legend()
plt.tight_layout()
plt.show()
```
